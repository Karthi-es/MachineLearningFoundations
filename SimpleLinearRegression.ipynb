# Simple Linear Regression

## Import the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,0:1].values
y = dataset.iloc[:,1].values
```

## Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Training simple linear regression model on the training dataset

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()    #This statement will create a simple linear regression model
regressor.fit(X_train, y_train)
```

## Predicting the test set results

```python
y_predict = regressor.predict(X_test)
```

Check the actual and predicted values of test set

```python
print(f'Actual values of test set: {y_test}')
print(f'Predicted values of test set: {y_predict}')
```

##Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color='yellow')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

##Visualising the test set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color='yellow')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```
